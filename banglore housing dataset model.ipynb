```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import math
```

```python
df=pd.read_csv(r"D:\harry bhai\pandas and numpy\Bengaluru_House_Data.csv")
```

```python
df
```

```python
df["size"].isnull().value_counts()
```

```python
df[["Room","Kithall"]]=pd.DataFrame(df['size'].str.split(' ',n=1,expand=True))
```

```python
df['Kithall'].replace("BHK",2,inplace=True)
df['Kithall'].replace("Bedroom",0,inplace=True)
df['Kithall'].replace("RK",0,inplace=True)
df["Room"]=df["Room"].astype(float)
df["total_size"]=df["Room"] + df["Kithall"]
```

```python
df
```

```python
df.drop(["Room"],axis=1,inplace=True)
df.drop(["Kithall"],axis=1,inplace=True)
```

```python
bath_mean=df["bath"].mean()
bath_mean
df["bath"].fillna(bath_mean,inplace=True)
df["bath"].isnull().value_counts().count
```

```python
df["total_size"].isnull().value_counts()
df["total_size"].fillna(df['total_size'].mean(),inplace=True)
df["balcony"].fillna(df["balcony"].mean(),inplace=True)
df
```

```python
df[['sq1','sq2']]=pd.DataFrame(df['total_ Sqft'].str.split(' ',n=1,expand=True))
```

```python
df
```

```python
def convertval(var):
    total=0
    if '-' in var:
        var1=var.split('-')
        for a in var1:
            total=total+float(a)
        avg=total/len(var1)
        return avg
    else:
        return float(var)
df["sq1"]=df["sq1"].apply(convertval)
```

```python
df
```

```python
df['sq2'].replace("Sq.Meter",10.7639,inplace=True)
df['sq2'].replace("Sq.Yards",9,inplace=True)
df['sq2'].replace("Acres",43560,inplace=True)
df['sq2'].replace("Grounds",2400,inplace=True)
df['sq2'].replace("Cents",435.6,inplace=True)
df['sq2'].replace("Guntha",1089,inplace=True)
df['sq2'].replace("Perch",272.25,inplace=True)
```

```python
df['sq2'].replace(np.NaN,1,inplace=True)
```

```python
df['sqft']=df['sq1']*df['sq2']
```

```python
df.drop(["sq1"],axis=1,inplace=True)
```

```python
df.drop(["sq2"],axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x=df.drop(['price','area_type',"availability",'location','size','society','total_ Sqft'],axis=1)
y=df["price"]
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.5,random_state=42)
x_train['bath'].size
```

```python
x_test['bath'].size
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[]])
```

```python
model.intercept_
```
